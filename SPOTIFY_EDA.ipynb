## Spotify is a digital music service that gives you access to millions of songs.

>The Spotify Song Attributes Dataset is a comprehensive collection of music tracks, encompassing various genres and artist names. This dataset provides valuable insights into the world of music, allowing enthusiasts, researchers, and data scientists to delve into the characteristics and nuances of each track.


>The dataset includes key features such as danceability, energy, loudness, speechiness, acousticness, instrumentalness, liveness, valence, tempo, duration, and time signature. These attributes provide a holistic view of the songs, enabling users to analyze and interpret different aspects of their musical composition.

```python
#importing the libraries 

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as stl
stl.use("ggplot")

#ignoring the warnings 
import warnings
warnings.filterwarnings("ignore")
```

```python
#importing the dataset

data=pd.read_csv('Spotify.csv')
data
```

- CLEANING THE DATA :-

```python
#to Check for Null Value 

data.isnull().sum()
```

```python
#to Drop the rows with null values

data.dropna(inplace=True)
data
```

```python
# to display only the trackName with its danceability

data.groupby('trackName')['danceability'].sum().round().sort_values(ascending=True)
```

```python
# to create a bar graph

df=pd.DataFrame(data.tail())
x=list(df.iloc[:,1])
y=list(df.iloc[:,6])
plt.bar(x,y,color='g')
plt.title('BAR GRAPH')
plt.xlabel('trackName')
plt.ylabel('energy')

plt.show()
```

- OBSERVATIONS :-

->  First and the last five entries are displayed using head and tail.

->  The total number of attributes in this dataset is 22 and the total number of entries are 10080.

->  No duplicate column is found in the dataset.

->  Rows and Columns were transposed using transform() function. 

->  After dropping the null values the number of entries reduces to 8580.

->  The above bar graph shows the energy of 5 particular tracks from the bottom of the dataset.

- 10 MOST POPULAR SONGS ON SPOTIFY

```python
#to display the top 10 most popular songs (tracks) on spotify

most = data.query('liveness > 0.4', inplace = False).sort_values('liveness', ascending = False)
most[:10]
```

##### OBV -  These are the most liked and popular songs on spotify.

- BAR PLOT

```python
# to print the duration of the songs in different genere 

plt.title('Duration of songs in different Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='trackName', x='duration_ms', data=data.head(20))
plt.xlabel('Duration in milliseconds')
plt.ylabel('track')
```

##### OBV - (Ijust) died in your arms is the longest song from the first 20 entries of the spotify dataset.

- REG PLOT

```python
# to print the correlation between speechiness and liveness 

plt.figure(figsize=(10,6))
sns.regplot(data=data, y='speechiness', x='liveness', color='g').set(title='Speechiness vs Liveness Correlation')
```

##### OBV - Correlation between speechiness and liveness is densly populated.

- TOP 5 GENRE 

```python
#to print the top 5 genre by liveness 

sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
popular = data.sort_values('liveness', ascending=False).head(10)
sns.barplot(y = 'genre', x = 'liveness', data = popular).set(title='Top 5 Genres by liveness :')
```

##### OBV - These are the top 5 genre by liveness

- BOX PLOT

```python
# to print the outliers in the tempo attribute 

sns.boxplot(data['tempo'])
```

##### OBV - Outliers are present in the 'tepmo' attribute of the dataset.(normal rangeis from 40 to 205)

- REL PLOT

```python
# Basically relplot() gives us scatterplot() and if we use kind as scatter , it gives us the scattr plot

sns.set(style ="ticks")
#tips = sns.load_dataset('data')
 
sns.relplot(x ="loudness",
            y ="tempo",
            kind ="scatter",
            data = data)
```

##### OBV - Rel plot shows the distribution of the data between 'loudness' and 'tempo' attributes on the dataset.

- HEATMAP

```python
# to print the pairewise correlation of all columns in the database

print(data.corr())

# to print the heatmap 

plt.figure(figsize=(10,8))
print(sns.heatmap(data.corr(),annot=True,cmap='Greens_r'))
```

##### OBV - Correlation between the attributes and the heatmap is shown .
